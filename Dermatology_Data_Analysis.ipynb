```python
import pandas as pd
```

```python
ds=pd.read_csv('dermatology_database_1.csv')
ds
```

```python
y=ds['class']
y
```

```python
x=ds.drop('class',axis=1)
x
```

```python
#Data Spliiting
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100 )
```

```python
x_train
```

```python
x_test
```

```python
#Model - Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_lr_train_prediction = lr.predict(x_train)
y_lr_test_prediction = lr.predict(x_test)
```

```python
print(y_lr_train_prediction)
```

```python
print(y_lr_test_prediction)
```

```python
#Model Performance
```

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train,y_lr_train_prediction)
lr_train_r2 = r2_score(y_train, y_lr_train_prediction)

lr_test_mse = mean_squared_error(y_test, y_lr_test_prediction)
lr_test_r2 = r2_score(y_test, y_lr_test_prediction)

print(f'R-squared (RÂ²) accuracy: {lr_test_r2}')
```

```python
print('LR MSE (Train) =',lr_train_mse)
print('LR R2 (Train) =', lr_train_r2)
print('LR MSE (Test) =' ,lr_test_mse)
print('LR R2 (Test) = ', lr_test_r2)
```

```python
#Model - Random Forest
```

```python
from sklearn.ensemble import RandomForestRegressor 

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

```python
y_rf_train_prediction = rf.predict(x_train)
y_rf_test_prediction = rf.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train,y_rf_train_prediction)
rf_train_r2 = r2_score(y_train, y_rf_train_prediction)

rf_test_mse = mean_squared_error(y_test, y_rf_test_prediction)
rf_test_r2 = r2_score(y_test, y_rf_test_prediction)
```

```python
print('RF MSE (Train) =',rf_train_mse)
print('RF R2 (Train) =', rf_train_r2)
print('RF MSE (Test) =' ,rf_test_mse)
print('RF R2 (Test) = ', rf_test_r2)

print(f'R-squared (RÂ²) accuracy: {rf_test_r2}')
```

```python
#Model Comparision
```

```python
print('LR MSE (Train) =',lr_train_mse)
print('LR R2 (Train) =', lr_train_r2)
print('LR MSE (Test) =' ,lr_test_mse)
print('LR R2 (Test) = ', lr_test_r2)

print('RF MSE (Train) =',rf_train_mse)
print('RF R2 (Train) =', rf_train_r2)
print('RF MSE (Test) =' ,rf_test_mse)
print('RF R2 (Test) = ', rf_test_r2)
```

```python
#Data Visualization of Result
```

```python
import matplotlib.pyplot as plt
plt.scatter(x=y_train, y=y_lr_train_prediction, c='brown', alpha=0.3)
plt.plot

plt.ylabel(' Predict class')
plt.xlabel(' Experimental class')
```

```python
#Model - Support Vector Regressor Model
```

```python
from sklearn.svm import SVR

svr_model = SVR()

svr_model.fit(x_train, y_train)
```

```python
svr_y_pred = svr_model.predict(x_test)
```

```python
svr_mse = mean_squared_error(y_test, svr_y_pred)
svr_r2 = r2_score(y_test, svr_y_pred)

print("\nSupport Vector Regressor Model:")
print(f'Mean Squared Error (MSE): {svr_mse}')
print(f'R-squared (RÂ²) accuracy: {svr_r2}')
```

```python
#Model- Gradient Boosting Model
```

```python
#Training the model

from sklearn.ensemble import GradientBoostingRegressor

gb_model = GradientBoostingRegressor(n_estimators=100, random_state=42)

gb_model.fit(x_train, y_train)
```

```python
#Model Prediction
gb_y_pred = gb_model.predict(x_test)
```

```python
#Model Performance
```

```python
gb_mse = mean_squared_error(y_test, gb_y_pred)
gb_r2 = r2_score(y_test, gb_y_pred)

print("\nGradient Boosting Model:")
print(f'Mean Squared Error (MSE): {gb_mse}')
print(f'R-squared (RÂ²) accuracy: {gb_r2}')
```
